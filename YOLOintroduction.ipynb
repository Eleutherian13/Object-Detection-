```python
img = show_img("pexels-photo-13872248.jpeg")
```

```python
# python usage 
```

```python
from ultralytics import YOLO


# reading from the documents 

model = YOLO("yolov8n.pt")  # load an official model
```

```python
results = model(img)
```

```python
# accessing the results 
```

```python

for i , results in enumerate(results):

    xywh = results[0].boxes[0].xywh
    print("Bounding box (xywh)")
    print(xywh)

    xywhn = results[0].boxes[0].xywhn  # normalized
    print("  Normalized Bounding Boxes (xywhn):")
    print(xywhn)

    xyxy = results[0].boxes[0].xyxy  # top-left-x, top-left-y, bottom-right-x, bottom-right-y
    print("  Bounding Boxes (xyxy):")
    print(xyxy)

    xyxyn = results[0].boxes[0].xyxyn  # normalized
    print("  Normalized Bounding Boxes (xyxyn):")
    print(xyxyn)

    names = [results[0].names[cls.item()] for cls in results[0].boxes[0].cls.int()]  # class name of each box
    print("  Class Names:")
    print(names)

    confs = results[0].boxes[0].conf  # confidence score of each box
    print("  Confidence Scores:")
    print(confs)
```
